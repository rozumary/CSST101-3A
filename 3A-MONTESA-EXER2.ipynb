**Objective:** Understand the fundamentals of probability theory and its application in AI for decision-making under uncertainty

Task:
1. Basic Probability Calculation: Write Python functions that calculate the following: Joint, Marginal and Conditional Probability

```python
def joint_probability(prob_A, prob_B):
    return prob_A * prob_B

def marginal_probability(joint_probs):
    return sum(joint_probs)

def conditional_probability(prob_A_given_B, prob_B):
    return prob_A_given_B * prob_B

prob_A = 0.3
prob_B = 0.4
prob_A_given_B = 0.5

joint_prob = joint_probability(prob_A, prob_B)
print(f"Joint Probability of A and B: {joint_prob}")

joint_probs = [0.12, 0.18, 0.08]
marginal_prob = marginal_probability(joint_probs)
print(f"Marginal Probability: {marginal_prob}")

conditional_prob = conditional_probability(prob_A_given_B, prob_B)
print(f"Conditional Probability of A given B: {conditional_prob}")
```

# **Exercise 2: Decision-Making Under Uncertainty**

**Objective:** Apply probability theory to make decisions in uncertain environments, such as AI systems that need to weigh the outcomes of different choices.

Task 1: Simulate a Decision Process

```python
import random

prob_success = 0.7
prob_failure = 1 - prob_success


reward = 100
cost = -50

expected_value = (prob_success * reward) + (prob_failure * cost)

print(f"Expected Value of making the decision: {expected_value}")
```

Decision Tree

```python
move_forward_prob_success = 0.6
move_forward_reward = 150
move_forward_cost = -100

turn_left_prob_success = 0.8
turn_left_reward = 80
turn_left_cost = -30


ev_move_forward = (move_forward_prob_success * move_forward_reward) + \
                  ((1 - move_forward_prob_success) * move_forward_cost)

ev_turn_left = (turn_left_prob_success * turn_left_reward) + \
               ((1 - turn_left_prob_success) * turn_left_cost)

best_decision = "Move forward" if ev_move_forward > ev_turn_left else "Turn left"

print(f"Expected Value for moving forward: {ev_move_forward}")
print(f"Expected Value for turning left: {ev_turn_left}")
print(f"The best decision is: {best_decision}")
```

# **Assessment Task 2: Bayesian Inference**

1. Implement a Bayesian Inference Function

2. Interactive Scenario:
o Scenario: A test for a disease has a 90% accuracy for positive results, and 5% of people without the disease test positive. If 1% of the population has the disease, what is the probability that someone who tested positive actually has the disease?
3. Feedback

```python
def bayesian_inference(p_h, p_e_given_h, p_e_given_not_h, p_not_h):
    p_e = (p_e_given_h * p_h) + (p_e_given_not_h * p_not_h)
    p_h_given_e = (p_e_given_h * p_h) / p_e
    return p_h_given_e

p_h = 0.01
p_e_given_h = 0.9
p_e_given_not_h = 0.05
p_not_h = 1 - p_h

probability_disease_given_positive = bayesian_inference(p_h, p_e_given_h, p_e_given_not_h, p_not_h)

print(f"The probability of having the disease given a positive test result is: {probability_disease_given_positive:.4f}")
```

# **Exercise 3: Applying Probability Theory in AI for Diagnosis**
**Objective:** Use probabilistic reasoning to model uncertainty in a real-world scenario such as medical diagnosis.

Task: Medical Diagnosis

```python
def calculate_posterior_probability(prior_disease_prob, sensitivity, false_positive_rate, prior_no_disease_prob):
    total_probability_positive_test = (sensitivity * prior_disease_prob) + (false_positive_rate * prior_no_disease_prob)
    posterior_probability_disease_given_positive = (sensitivity * prior_disease_prob) / total_probability_positive_test
    return posterior_probability_disease_given_positive

prior_disease_probability = 0.02
sensitivity_of_test = 0.95
false_positive_rate_of_test = 0.1
prior_no_disease_probability = 1 - prior_disease_probability

posterior_probability = calculate_posterior_probability(prior_disease_probability, sensitivity_of_test, false_positive_rate_of_test, prior_no_disease_probability)

print(f"The probability of having the disease given a positive test result is: {posterior_probability:.4f}")
```

# **Assessment Task 3: Decision-Making Under Uncertainty**
1. Interactive Decision Simulation:
oYou are tasked with simulating a decision-making process using Python. In this simulation, an AI system must choose between two investment options with uncertain outcomes (profit or loss).
2.Interactive Task: Implement a function that simulates the decision-making process for 1000 iterations. The probability of success is 70%, and the reward for success is $1000, while the penalty for failure is -$500.
o Interactive Input: Adjust the probability of success or failure and simulate different outcomes.

3. Question: What is the average return after 1000 iterations? How does changing the probability affect the result?
4. Feedback

```python
import random

def simulate_investment_decision(probability_of_success, reward, penalty, iterations):
    total_return = 0
    for _ in range(iterations):
        if random.random() < probability_of_success:
            total_return += reward
        else:
            total_return += penalty
    average_return = total_return / iterations
    return average_return

# parameters
probability_of_success = 0.7
reward_for_success = 1000
penalty_for_failure = -500
num_iterations = 1000

average_return = simulate_investment_decision(probability_of_success, reward_for_success, penalty_for_failure, num_iterations)

print(f"Average return after {num_iterations} iterations: ${average_return:.2f}")
```

# **Exercise 4: Probability Distribution in AI**
**Objective:** Work with probability distributions that are key in AI for representing uncertainty.
Tasks:

1. Simulate Probability Distributions: Write a Python script that generates data from common probability distributions (e.g., binomial, normal) using numpy or scipy.

2. Visualize the Distribution: Use a plotting library (e.g., matplotlib) to visualize the distribution and interpret the results.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def simulate_binomial_distribution(n, p, size):
    return np.random.binomial(n, p, size)

def simulate_normal_distribution(mean, std_dev, size):
    return np.random.normal(mean, std_dev, size)

def plot_distribution(data, distribution_type):
    plt.figure(figsize=(10, 6))
    sns.histplot(data, bins=30, kde=True, color=sns.color_palette("plasma", 1)[0])
    plt.title(f'{distribution_type} Distribution')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.grid()
    plt.show()

# binomial
n_binomial = 10
p_binomial = 0.5
size_binomial = 1000

# normal
mean_normal = 0
std_dev_normal = 1
size_normal = 1000

binomial_data = simulate_binomial_distribution(n_binomial, p_binomial, size_binomial)
normal_data = simulate_normal_distribution(mean_normal, std_dev_normal, size_normal)

plot_distribution(binomial_data, 'Binomial')
plot_distribution(normal_data, 'Normal')
```

# **Assessment Task 4: Real-World Scenario**
1. Interactive Real-World Problem: You are asked to model a real-world scenario in AI, such as predicting whether it will rain based on factors like humidity and cloud cover. Use conditional probabilities to model this scenario.
2. Scenario Simulation Task: Implement a Python script that uses conditional probability to predict the likelihood of rain given evidence (humidity and cloud cover).

  Interactive Scenario: You can change the input conditions (e.g., increase humidity or cloud cover) to see how the probability of rain changes in real-time.
3. Feedback

*Rain Prediction Model*

```python
import numpy as np
import matplotlib.pyplot as plt

def conditional_probability_rain_given_evidence(p_rain, p_humidity_given_rain, p_cloud_cover_given_rain, p_humidity, p_cloud_cover):
    p_evidence_given_rain = p_humidity_given_rain * p_cloud_cover_given_rain
    p_evidence = p_humidity * p_cloud_cover
    p_rain_given_evidence = (p_evidence_given_rain * p_rain) / p_evidence
    return p_rain_given_evidence

def plot_prediction(humidity_values, cloud_cover_values, probabilities):
    plt.figure(figsize=(10, 6))
    plt.plot(humidity_values, probabilities, label='Rain Probability based on Humidity')
    plt.plot(cloud_cover_values, probabilities, label='Rain Probability based on Cloud Cover', linestyle='--')
    plt.xlabel('Humidity / Cloud Cover (%)')
    plt.ylabel('Probability of Rain')
    plt.title('Probability of Rain based on Evidence (Humidity and Cloud Cover)')
    plt.legend()
    plt.grid(True)
    plt.show()

p_rain = 0.3
p_humidity_given_rain = 0.8
p_cloud_cover_given_rain = 0.9
p_humidity = 0.6
p_cloud_cover = 0.7

humidity_values = np.linspace(0, 1, 100)
cloud_cover_values = np.linspace(0, 1, 100)
probabilities = [conditional_probability_rain_given_evidence(p_rain, h, c, p_humidity, p_cloud_cover) for h, c in zip(humidity_values, cloud_cover_values)]

plot_prediction(humidity_values, cloud_cover_values, probabilities)
```

# **Exercise 5: Real-World Application of Probability in AI**
**Objective:** Apply probability theory to real-world AI scenarios, such as predicting customer behavior or making decisions in uncertain markets.

Tasks:

1.Customer Behavior Prediction: Suppose you are working for an e-commerce company and need to predict the probability that a customer will make a purchase based on previous behavior.

Define probabilities for different user actions (e.g., clicking on a product, adding to cart, making a purchase).

Use conditional probability to predict the likelihood of a purchase given that a customer added an item to their cart.

```python
import numpy as np

def conditional_probability_purchase_given_cart(p_purchase, p_add_to_cart_given_purchase, p_add_to_cart):
    p_cart_given_purchase = p_add_to_cart_given_purchase * p_purchase
    p_purchase_given_cart = p_cart_given_purchase / p_add_to_cart
    return p_purchase_given_cart

p_purchase = 0.2
p_add_to_cart_given_purchase = 0.7
p_add_to_cart = 0.5

p_purchase_given_cart = conditional_probability_purchase_given_cart(p_purchase, p_add_to_cart_given_purchase, p_add_to_cart)

print(f"Probability of purchase given that the item was added to cart: {p_purchase_given_cart:.2f}")
```
