#**Part 2. Basic Probability Calculation**  
1. **Data:**  

Prior probability (prevalence of the disease):  
P(Disease) = 0.08  

Test sensitivity (probability of a positive result given disease):  
P(Positive | Disease) = 0.90  

False positive rate (probability of a positive result given no disease):  
P(Positive | NoDisease) = 0.07  

Probability of a positive result in the population:  
P(Positive) = 0.12  

2. **Computation**  

Substituting values into Bayes' Theorem Formula:  
P(Disease | Positive) = P(Positive | Disease) × P(Disease) / P(Positive)  

Insert the values:  
P(Disease | Positive) = 0.90 × 0.08 / 0.12  

0.90 × 0.08 = 0.072  

P(Disease | Positive) = 0.072 / 0.12  

P(Disease | Positive) = 0.60 (approximately)  

3.  **Findings:**

Posterior Probability: The posterior probability that the patient has asthma given a positive test result is approximately **0.60** or **60%.  **

#**Part 3. Decision-Making Simulation**  
1. **Data:**  

Probability of Success (P(Success)) = 75% = 0.75  

Benefit from Success = 80 points  

Probability of Failure (P(Failure)) = 25% = 0.25  

Loss from Failure = -30 points  

2. **Computation:**  

Expected Health Improvement = (0.75 × 80) + (0.25 × (-30))  

3. **Findings:**

Expected Improvement: The expected health improvement for the patient is **52.5 points.**  

#**Part 4:  Visualize a Probability Distribution**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

n = 6
p = 0.75

x = np.arange(0, n + 1)

probabilities = binom.pmf(x, n, p)

plt.figure(figsize=(10, 6))
plt.bar(x, probabilities, color='lightgreen', alpha=0.7)
plt.xticks(x)
plt.xlabel('Number of Successes', fontsize=14)
plt.ylabel('Probability', fontsize=14)
plt.title('Binomial Distribution: Probability of Success in 6 Trials', fontsize=16)
plt.ylim(0, max(probabilities) + 0.05)
plt.grid(axis='y', linestyle='--', alpha=0.7)

for i in range(len(x)):
    plt.text(x[i], probabilities[i], f'{probabilities[i]:.2f}', ha='center', va='bottom')

plt.show()
```
